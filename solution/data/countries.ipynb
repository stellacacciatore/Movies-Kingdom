# COUNTRIES.CSV CLEANING

```python
import pandas as pd
```

```python
data_directory = 'data_csv/'

countries_df = pd.read_csv(data_directory + 'countries.csv' ,quotechar='"')
countries_df
```

```python
countries_df.shape
```

```python
countries_df.dtypes
```

## CHANGING TYPES OF OBJECT COLUMNS

```python
countries_df['country'] = countries_df['country'].astype('string')
countries_df.dtypes
```

## NOW WE TRY TO UNDERSTAND COLUMNS VALUES


```python
print(list(countries_df.columns.values))
```

```python
list_of_countries = list(countries_df['country'].unique())
list_of_countries.sort()
print(list_of_countries)
```

```python
countries_df.rename(columns={'id':'movie_id'}, inplace=True)
print(list(countries_df.columns.values))
```

## ARE THERE NA VALUES IN THE DATASET?


```python
not_available_movie_ids = countries_df['movie_id'].isna().sum()
not_available_countries = countries_df['country'].isna().sum()
print('Not available movie ids: {}'.format(not_available_movie_ids),
      '\nNot available countries: {}'.format(not_available_countries))
```

## ARE THERE ANY DUPLICATES

```python
num_duplicated_countries = countries_df.duplicated().sum()
print('There are a total of {} duplicates '.format(num_duplicated_countries))
```

```python
countries_df.to_csv('data_cleaned/countries_cleaned.csv', index=False)
```
